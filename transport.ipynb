```python
plt.figure(figsize=(8, 8))
travel_details['Goods'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl', len(travel_details['Goods'].unique())))
plt.title('Goods Distribution by Occurence')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
travel_details['Transport Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl', len(travel_details['Transport Type'].unique())))
plt.title('Transport Types Distribution by Occurence')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
travel_details['Vendor ID'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl', len(travel_details['Vendor ID'].unique())))
plt.title('Total Supply Distribution by Vendors(Occurence)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Calculate the sum of actual costs for each goods type
actual_cost_sum = travel_details.groupby('Goods')['Actual Cost'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
bars=plt.bar(actual_cost_sum['Goods'], actual_cost_sum['Actual Cost'], color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
plt.title('Sum of Actual Costs by Goods Type')
plt.xlabel('Goods Type')
plt.ylabel('Sum of Actual Costs(in crores)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group the data by goods type and calculate the sum of actual and committed quantities
quantity_sum = travel_details.groupby('Goods')[['Committed Quantity', 'Actual Quantity']].sum()

# Get goods types and corresponding quantities
goods_types = quantity_sum.index
committed_quantity = quantity_sum['Committed Quantity']
actual_quantity = quantity_sum['Actual Quantity']

# Generate indices for x-axis
indices = np.arange(len(goods_types))

# Plot the bar chart
plt.figure(figsize=(14, 6))
bar_width = 0.35
bars1=plt.bar(indices, committed_quantity/1000, bar_width, label='Committed Quantity', color='skyblue')
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
bars2=plt.bar(indices + bar_width, actual_quantity/1000, bar_width, label='Actual Quantity', color='orange')
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')

plt.xlabel('Goods Type')
plt.ylabel('Quantity(in tonnes)')
plt.title('Comparison between Committed and Actual Quantities by Goods Type')
plt.xticks(indices + bar_width / 2, goods_types)
plt.legend()
plt.show()
```

```python
# Group the data by goods type and calculate the sum of actual and committed quantities
quantity_sum = travel_details.groupby('Goods')[['Committed Cost', 'Actual Cost']].sum()

# Get goods types and corresponding quantities
goods_types = quantity_sum.index
committed_cost = quantity_sum['Committed Cost']
actual_cost = quantity_sum['Actual Cost']

# Generate indices for x-axis
indices = np.arange(len(goods_types))

# Plot the bar chart
plt.figure(figsize=(14, 6))
bar_width = 0.35
bars1=plt.bar(indices, committed_cost, bar_width, label='Committed Cost', color='skyblue')
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
bars2=plt.bar(indices + bar_width, actual_cost, bar_width, label='Actual Cost', color='orange')
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')

plt.xlabel('Goods Type')
plt.ylabel('Cost(in crores)')
plt.title('Comparison between Committed and Actual Costs by Goods Type')
plt.xticks(indices + bar_width / 2, goods_types)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
cumulative_data = travel_details[['Committed Quantity', 'Actual Quantity']].cumsum()
cumulative_data.plot(kind='area', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title('Cumulative Committed and Actual Quantities Over Time')
plt.xlabel('Transaction')
plt.ylabel('Cumulative Quantity(in tonnes)')
plt.tight_layout()
plt.show()
```

```python
avg_delay_by_transport = travel_details.groupby('Transport Type')['Delay'].mean().reset_index()


# Plot bar plot
plt.figure(figsize=(10, 6))
ax=sns.barplot(x='Transport Type', y='Delay', data=avg_delay_by_transport, palette='husl')
plt.title('Average Delay Time by Transport Type')
plt.xlabel('Transport Type')
plt.ylabel('Average Delay(in days) Time')
plt.xticks(rotation=45)
plt.tight_layout()
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.tight_layout()
plt.show()
plt.show()
```

```python
avg_delay_by_vendor = travel_details.groupby('Vendor ID')['Delay'].mean().reset_index()


# Plot bar plot
plt.figure(figsize=(10, 6))
ax=sns.barplot(x='Vendor ID', y='Delay', data=avg_delay_by_vendor, palette='husl')
plt.title('Average Delay Time by Vendor')
plt.xlabel('Vendor ID')
plt.ylabel('Average Delay Time')
plt.xticks(rotation=45)
plt.tight_layout()
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.tight_layout()
plt.show()
plt.show()
```

```python
# Group the data by month and calculate the sum of actual quantities
actual_quantity_by_month = travel_details.groupby(travel_details['Month'])['Actual Quantity'].sum()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(actual_quantity_by_month.index, actual_quantity_by_month.values, marker='o', color='blue', linestyle='-')
plt.title('Actual Quantity Generated by Month')
plt.xlabel('Month')
plt.ylabel('Actual Quantity(in tonnes)')

# Set labels for the month in the form of Jan, Feb, Mar, etc.
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(range(1, 13), months)
for group,x,y in zip(actual_quantity_by_month,actual_quantity_by_month.index, actual_quantity_by_month.values):
    plt.text(x,y,group,va="bottom",ha="right",color="blue",fontsize=6)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group the data by month and calculate the sum of actual quantities
actual_quantity_by_month = travel_details.groupby(travel_details['Quarter'])['Actual Quantity'].sum()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(actual_quantity_by_month.index, actual_quantity_by_month.values, marker='o', color='blue', linestyle='-')
plt.title('Actual Quantity Generated by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Actual Quantity')

# Set labels for the month in the form of Jan, Feb, Mar, etc.
quarter = ['Q1','Q2','Q3','Q4']
plt.xticks(range(1, 5),quarter)
for group,x,y in zip(actual_quantity_by_month,actual_quantity_by_month.index, actual_quantity_by_month.values):
    plt.text(x,y,group,va="bottom",ha="right",color="blue",fontsize=6)
plt.grid(True)
plt.tight_layout()
plt.show()
```
