```python
import pandas as pd
import numpy as np
```

```python
num_rows=1000

#security metrics
physical_security_metrics=np.random.randint(3,6,size=num_rows)
cybersecurity_metrics=np.random.randint(3,6,size=num_rows)
regulatory_compliance_metrics=np.random.randint(2,6,size=num_rows)

#traceability metrics
product_tracking_metrics=np.random.randint(2,6,size=num_rows)
documentation_metrics=np.random.randint(3,6,size=num_rows)
transparency_metrics=np.random.randint(2,6,size=num_rows)

# Create the dataset
data = {
    'Physical Security Metrics': physical_security_metrics,
    'Cyber Security Metrics': cybersecurity_metrics,
    'Regulatory Compliance Metrics': regulatory_compliance_metrics,
    'Product Tracking Metrics': product_tracking_metrics,
    'Documentation Metrics': documentation_metrics,
    'Transparency Metrics': transparency_metrics,
}

sec_trac=pd.DataFrame(data)
```

```python
sec_trac['Security_Score']=sec_trac['Physical Security Metrics']/5+sec_trac['Cyber Security Metrics']/5+sec_trac['Regulatory Compliance Metrics']/5
sec_trac['Traceability_Score']=sec_trac['Product Tracking Metrics']/5+sec_trac['Documentation Metrics']/5+ sec_trac['Transparency Metrics']/5
```

```python
travel_details=pd.read_csv('final_transport.csv')
```

```python
# Merge relevant columns from travel_details and energy_supply_chain_df
supply_df = pd.merge(travel_details[['Transport Type', 'Delay', 'Actual Timeline', 'Actual Cost']],
                     sec_trac[['Security_Score', 'Traceability_Score']],
                     left_index=True, right_index=True)

# Generate random values for Demand Forecast and Supply Chain Visibility
supply_df['Demand_Forecast'] = np.random.randint(6000, 10000, size=len(supply_df))
supply_df['Supply_Chain_Visibility'] = np.random.randint(1, 4, size=len(supply_df))  # Assuming a scale of 1 to 10
```

```python
supply_df
```

```python
supply_df['Supply_Chain_Visibility'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_values = {'Roadways': 1, 'Railways': 2, 'Waterways': 3}
label_encoder=LabelEncoder()
supply_df['Transport_Encoded'] = label_encoder.fit_transform(supply_df['Transport Type'].map(encoding_values))+1
```

```python
supply_df
```

```python
supply_df['On-Time Delivery Rate']=supply_df['Transport_Encoded']/3+(4-supply_df['Delay'])/4+(3-supply_df['Security_Score'])/3+(3-supply_df['Traceability_Score'])/3+(3-supply_df['Supply_Chain_Visibility'])/3+(9999-supply_df['Demand_Forecast'])/9999
```

```python
supply_df
```

```python
supply_df.describe()
```

```python
max=supply_df['On-Time Delivery Rate'].max()
supply_df['On-Time Delivery Rate']=supply_df['On-Time Delivery Rate']/max
```

```python
supply_df['On-Time Delivery Rate']*=100
```
