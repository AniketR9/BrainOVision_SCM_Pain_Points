```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vendor_df=pd.read_csv('vendor.csv')
```

```python
vendor_df
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
bars=plt.bar(vendor_df['Vendor ID'], vendor_df['Total Revenue'], color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
plt.title('Total Revenue generated by Vendor')
plt.xlabel('Vendor ID')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
vendor_no=vendor_df['Vendor ID']

indices=np.arange(len(vendor_no))

plt.figure(figsize=(12, 6))
barwidth=0.35
bars1=plt.bar(indices, vendor_df['Total Volume']/1000,barwidth, label='Total Volume', color='skyblue')
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
bars2=plt.bar(indices+barwidth, vendor_df['Total Actual Quantity']/1000,barwidth, label='Total Actual Quantity', color='orange')
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')
plt.xlabel('Vendor ID')
plt.ylabel('Quantity(in tonnes)')
plt.title('Comparison between Total Volume and Total Actual Quantity by Vendor ID')
plt.xticks(indices+barwidth/2,vendor_no)
plt.legend()
plt.tight_layout()
plt.show()
```
